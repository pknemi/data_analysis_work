```python
import pandas as pd
```

```python
df=pd.read_csv('pandas_docs/loan_data.csv')
df
```

# Q1- If the loan amount is above 100,check whether they are self employed or not and their applicant income

```python
df[df['LoanAmount']<100].loc[:,['LoanAmount','Self_Employed','ApplicantIncome']]
```

# Q2- Print all data of those who are having below 3000 as applicant income

```python
df[df['ApplicantIncome']<3000]
```

# Q3- find the average of applicant income?

```python
df['ApplicantIncome'].mean()
```

# Q4- find the percentage of graduates and non graduates among those who applied for loan?

```python
df['Education'].value_counts(normalize=True)
```

# Q5- what is the maximum value of applicant income and find what is the loan amount he got?

```python
df[df['ApplicantIncome']==df['ApplicantIncome'].max()].loc[:,['Loan_ID','ApplicantIncome','LoanAmount']]
```

# Q6- what is the minimum value of the applicant income and find what is his loan amount?

```python
df[df['ApplicantIncome']==df['ApplicantIncome'].min()].loc[:,['Loan_ID','ApplicantIncome','LoanAmount']]
```

# Q7- what is the maximum loan amount issued? find his education, marital status, property area, applicant income and gender?

```python
df[df['LoanAmount']==df['LoanAmount'].max()].loc[:,['Loan_ID','ApplicantIncome','LoanAmount','Gender','Education','Property_Area']]
```

# Q8- if the gender of the applicant is male and graduated, find his applicant income, loan amount and loan status?

```python
df[(df['Gender']=='Male') & (df['Education']=='Graduate')].loc[:,['Gender','Education','ApplicantIncome','LoanAmount','Loan_Status']]
```

# Q9- if the gender is female, married and self employed, check her loan status, loan amount and loan term?

```python
df[(df['Gender']=='Female') & (df['Married']=='Yes') & (df['Self_Employed']=='Yes')].loc[:,['Gender','Married','LoanAmount','Loan_Status','Loan_Amount_Term']]
```

# Q10- Describe the whole data?

```python
df.describe()
```

# Q11- if the property area is rural or the gender is women, find their loan amount and applicant income?

```python
df[(df['Property_Area']=='Rural') | (df['Gender']=='Female')].loc[:,['LoanAmount','ApplicantIncome']]
```

# Q12- If the credit history is 1, what is the maximum and minimum loan amount provided?

```python
df[(df['Credit_History']==1) & (df['LoanAmount']==df['LoanAmount'].max())].loc[:,['Credit_History','LoanAmount']]
```

```python
df[(df['Credit_History']==1) &  (df['LoanAmount']==df['LoanAmount'].min())].loc[:,['Credit_History','LoanAmount']]
```

# Q13- If a person is Graduate but his credit History is 0, find his loan amount and its status?

```python
df[(df['Education']=='Graduate') & (df['Credit_History']==0)] .loc[:,['Education','Credit_History','LoanAmount','Loan_Status']].value_counts()
```

# Q14- find the applicant who has least loan lerm and maximum loan amount

```python
df[df['Loan_Amount_Term']==df['Loan_Amount_Term'].min()].loc[:,['Loan_Amount_Term','LoanAmount']]
```
